```python
import numpy as np 
import matplotlib.pyplot as plt
```

We are given the initial value problem:

$$
\frac{dy}{dt} = \cos(t) - y, \quad y(0) = 1
$$

solving using euler's method for n=20


```python
# Define the function f(t, y) = dy/dt
def f(t, y):
    return np.cos(t) - y
```

```python
# Parameters
a = 0       # Start of interval
b = 5       # End of interval
y0 = 1      # Initial condition y(0)
n = 20      # Number of steps
h = (b - a) / n  # Step size
```

```python
# Initialize arrays for t and y
t = np.linspace(a, b, n + 1)
y = np.zeros(n + 1)
y[0] = y0
```

```python
# Euler's method loop
for i in range(n):
    y[i + 1] = y[i] + h * f(t[i], y[i])
```

```python
# Plot the solution
plt.plot(t, y, 'o-', label="Euler's Method")
plt.xlabel('t')
plt.ylabel('y')
plt.title("Euler's Method Solution to y' = cos(t) - y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd

# Create a DataFrame from the t and y arrays
results_df = pd.DataFrame({
    't': t,
    'y_euler': y
})

# Save to CSV
results_df.to_csv('euler_solution.csv', index=False)
```

### Exact Solution of the IVP

We have:

$$
\frac{dy}{dt} = \cos(t) - y, \quad y(0) = 1
$$

This is a linear first-order ODE. To solve it, we use an **integrating factor**:

1. **Write the equation in standard linear form**:

$$
\frac{dy}{dt} + y = \cos(t)
$$

2. **Find the integrating factor** $\mu(t)$:

$$
\mu(t) = e^{\int 1\, dt} = e^t
$$

3. **Multiply both sides of the equation by** $e^t$:

$$
e^t \frac{dy}{dt} + e^t y = e^t \cos(t)
$$

$$
\frac{d}{dt}\left(e^t y\right) = e^t \cos(t)
$$

4. **Integrate both sides**:

$$
e^t y = \int e^t \cos(t)\, dt
$$

Use integration by parts or a known identity:

$$
\int e^t \cos(t)\, dt = \frac{e^t}{2} (\sin(t) + \cos(t)) + C
$$

So,

$$
e^t y = \frac{e^t}{2} (\sin(t) + \cos(t)) + C
$$

Divide both sides by $e^t$:

$$
y(t) = \frac{1}{2}(\sin(t) + \cos(t)) + Ce^{-t}
$$

5. **Apply initial condition** $y(0) = 1$:

$$
1 = \frac{1}{2}(0 + 1) + C \cdot 1 \Rightarrow C = \frac{1}{2}
$$

6. **Final exact solution**:

$$
\boxed{y(t) = \frac{1}{2}(\sin(t) + \cos(t)) + \frac{1}{2}e^{-t}}
$$


## plotting for n=100

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Define the differential equation dy/dt = cos(t) - y
def f(t, y):
    return np.cos(t) - y

# Exact solution
def exact_solution(t):
    return 0.5 * (np.sin(t) + np.cos(t)) + 0.5 * np.exp(-t)

# Parameters
a = 0       # Start of interval
b = 5       # End of interval
y0 = 1      # Initial condition y(0)
n = 100     # Number of steps
h = (b - a) / n  # Step size

# Initialize arrays
t = np.linspace(a, b, n + 1)
y_euler = np.zeros(n + 1)
y_euler[0] = y0

# Euler's method
for i in range(n):
    y_euler[i + 1] = y_euler[i] + h * f(t[i], y_euler[i])

# Compute exact solution
y_exact = exact_solution(t)

# Plotting both solutions
plt.plot(t, y_euler, 'o-', label=f"Euler's Method(n={n})", markersize=4)
plt.plot(t, y_exact, '-', label="Exact Solution", linewidth=2)
plt.xlabel('t')
plt.ylabel('y')
plt.title("Euler's Method vs Exact Solution")
plt.grid(True)
plt.legend()
plt.show()
plt.savefig('results(n=100)')
```
